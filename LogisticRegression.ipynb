```python
#importing supporting libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import random

#importing sklearn libraries for models
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
#importing data
train_df = pd.read_csv('sign_mnist_train.csv')
test_df = pd.read_csv('sign_mnist_test.csv')
```

```python
#Separating labels and data
y_train = train_df['label']
y_test = test_df['label']

del train_df['label']
del test_df['label']

X_train = train_df.values
X_test = test_df.values
```

```python
#Reshaping labels into 2D arrays
y_train = np.array(y_train).reshape((y_train.shape[0],1))
y_test = np.array(y_test).reshape((y_test.shape[0],1))
```

```python
#Normalization
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
#Checking how the images look like
for i in range(5):
    j = random.randrange(0,len(X_train),1)
    plt.imshow(X_train[j].reshape((28,28)), cmap=plt.get_cmap('gray'))
    plt.show()
```

```python
#Performing trasformations on data for further model attempts

X_train_sqr = X_train**2
X_test_sqr = X_test**2

X_train_cb = X_train**3
X_test_cb = X_test**3

all_X_train = [X_train, X_train_sqr, X_train_cb]
all_X_test = [X_test, X_test_sqr, X_test_cb]
```

```python
transformations = ['None', 'Squared', 'Cubed']
```

```python
#Logistic Regression
train_acc_logreg = []
test_acc_logreg = []
for (i,j,k) in zip(all_X_train, all_X_test, transformations):
    logreg = LogisticRegression(penalty = 'none')
    logreg.fit(i, y_train)
    test_score_logreg = logreg.score(j, y_test)
    train_score_logreg = logreg.score(i, y_train)
    train_acc_logreg.append(train_score_logreg)
    test_acc_logreg.append(test_score_logreg)
    print(k)
    print('Train Accuracy: ', train_score_logreg)
    print('Test Accuracy: ', test_score_logreg)

df_logreg = pd.DataFrame({"Transformation": transformations, "Training Accuracy": train_acc_logreg, "Testing Accuracy" : test_acc_logreg}, index = None) 
print(df_logreg)
    
```

```python
#Setting up C values to test on
C_list = [0.001, 0.01, 0.1, 1.0, 2.0, 5.0, 10.0]
```

```python
#Lasso Regression
train_acc_logreg1 = {'None': [], 'Squared': [], 'Cubed': []}
test_acc_logreg1 = {'None': [], 'Squared': [], 'Cubed': []}

for (i,j,k) in zip(all_X_train, all_X_test, transformations):
    l1_scores_test = []
    l1_scores_train = []
    for c in C_list:
        logreg1 = LogisticRegression(penalty = 'l1', C = c, solver = 'liblinear', tol = 0.1)
        logreg1.fit(i, y_train)
        test_score = logreg1.score(j, y_test)
        l1_scores_test.append(test_score)
        train_score = logreg1.score(i, y_train)
        l1_scores_train.append(train_score)

    plt.plot(C_list, l1_scores_train, color = 'blue', label = 'Train')
    plt.plot(C_list, l1_scores_test, color = 'red', label = 'Test')
    plt.title('Lasso Regression Results With Feature Transformation ' + k)
    plt.xlabel('C')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.legend()
    plt.show()

    train_acc_logreg1[k] = l1_scores_train
    test_acc_logreg1[k] = l1_scores_test

df_logreg1 = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg1.get('None'), 'Testing Accuracy': test_acc_logreg1.get('None')}, index = None)
df_logreg1_sqr = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg1.get('Squared'), 'Testing Accuracy': test_acc_logreg1.get('Squared')}, index = None)
df_logreg1_cb = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg1.get('Cubed'), 'Testing Accuracy': test_acc_logreg1.get('Cubed')}, index = None)

print('No Tranfromation \n', df_logreg1)
print('Squared Features \n', df_logreg1_sqr)
print('Cubed Features \n', df_logreg1_cb)
```

```python
#Ridge Regression
train_acc_logreg2 = {'None': [], 'Squared': [], 'Cubed': []}
test_acc_logreg2 = {'None': [], 'Squared': [], 'Cubed': []}

for (i,j,k) in zip(all_X_train, all_X_test, transformations):
    l2_scores_test = []
    l2_scores_train = []
    for c in C_list:
        logreg2 = LogisticRegression(penalty = 'l2', C = c, tol = 0.1)
        logreg2.fit(i, y_train)
        test_score = logreg2.score(j, y_test)
        l2_scores_test.append(test_score)
        train_score = logreg2.score(i, y_train)
        l2_scores_train.append(train_score)

    plt.plot(C_list, l2_scores_train, color = 'blue', label = 'Train')
    plt.plot(C_list, l2_scores_test, color = 'red', label = 'Test')
    plt.title('Ridge Regression Results With Feature Transformation ' + k)
    plt.xlabel('C')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.legend()
    plt.show()

    train_acc_logreg2[k] = l2_scores_train
    test_acc_logreg2[k] = l2_scores_test

df_logreg2 = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg2.get('None'), 'Testing Accuracy': test_acc_logreg2.get('None')}, index = None)
df_logreg2_sqr = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg2.get('Squared'), 'Testing Accuracy': test_acc_logreg2.get('Squared')}, index = None)
df_logreg2_cb = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg2.get('Cubed'), 'Testing Accuracy': test_acc_logreg2.get('Cubed')}, index = None)

print('No Tranfromation \n', df_logreg2)
print('Squared Features \n', df_logreg2_sqr)
print('Cubed Features \n', df_logreg2_cb)
```

```python
#Elasticnet Regression
train_acc_logreg_ela = {'None': [], 'Squared': [], 'Cubed': []}
test_acc_logreg_ela = {'None': [], 'Squared': [], 'Cubed': []}


for (i,j,k) in zip(all_X_train, all_X_test, transformations):
    ela_scores_test= []
    ela_scores_train = []
    for c in C_list:
        logreg3 = LogisticRegression(penalty = 'elasticnet', C = c, tol = 0.1, solver = 'saga', l1_ratio = 0.5)
        logreg3.fit(i, y_train)
        test_score = logreg3.score(j, y_test)
        ela_scores_test.append(test_score)
        train_score = logreg3.score(i, y_train)
        ela_scores_train.append(train_score)

    plt.plot(C_list, ela_scores_train, color = 'blue', label = 'Train')
    plt.plot(C_list, ela_scores_test, color = 'red', label = 'Test')
    plt.title('Elasticnet Regression With Feature Transformation ' + k)
    plt.xlabel('C')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.legend()
    plt.show()

    train_acc_logreg_ela[k] = ela_scores_train
    test_acc_logreg_ela[k] = ela_scores_test

df_logreg_ela = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg_ela.get('None'), 'Testing Accuracy': test_acc_logreg_ela.get('None')}, index = None)
df_logreg_ela_sqr = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg_ela.get('Squared'), 'Testing Accuracy': test_acc_logreg_ela.get('Squared')}, index = None)
df_logreg_ela_cb = pd.DataFrame({'C': C_list, 'Training Accuracy': train_acc_logreg_ela.get('Cubed'), 'Testing Accuracy': test_acc_logreg_ela.get('Cubed')}, index = None)

print('No Tranfromation \n', df_logreg_ela)
print('Squared Features \n', df_logreg_ela_sqr)
print('Cubed Features \n', df_logreg_ela_cb)
```
